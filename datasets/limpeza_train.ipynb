# Limpeza do arquivo train.csv

```python
import pandas as pd
import datetime

# criar o dataframe
df = pd.read_csv('train.csv')
df
```

```python
df.dtypes # Mostra os tipos de dados de cada coluna
            # object = texto / string
```

## Retira linhas com 'NaN'

```python

# seleciona apenas linhas que não tem 'NaN ' na coluna de idade, avalização, multiple_deliveries, densidade de tráfego, condição climatica e Cidade
linhas_com_valor = (df['Delivery_person_Age'] != 'NaN ') & (df['Delivery_person_Ratings'] != 'NaN ') & (df['multiple_deliveries'] != 'NaN ')
# Deixa apenas as linhas que não tem 'NaN ' em nenhuma das colunas citadas acima
df = df.loc[linhas_com_valor, :]

# seleciona apenas linhas que não tem 'NaN ' na coluna de idade, avalização, multiple_deliveries, densidade de tráfego, condição climatica e Cidade
linhas_com_valor = (df['Road_traffic_density'] != 'NaN ') & (df['Weatherconditions'] != 'conditions NaN') & (df['City'] != 'NaN ') & (df['Festival'] != 'NaN ')
# Deixa apenas as linhas que não tem 'NaN ' em nenhuma das colunas citadas acima
df = df.loc[linhas_com_valor, :]
```

## Converte o tipo de algumas colunas que está errado

```python
# Conversões:
df['Delivery_person_Age'] = df['Delivery_person_Age'].astype(int) # converte a coluna de idade de object para inteiro

df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].astype(float) # muda o tipo da coluna Delivery_person_Ratings que está como Object e deve ser Float

# Usa a biblioteca datetime, importada na primeira célula
df['Order_Date'] = pd.to_datetime( df['Order_Date'], format='%d-%m-%Y' ) # Corrigindo a coluna Order_Date. Para mudar para data é diferente, precisamos importar ima biblioteca

df['multiple_deliveries'] = df['multiple_deliveries'].astype(int) # converte a coluna para inteiro

df['Time_taken(min)'] = df['Time_taken(min)'].str.replace('(min) ', '').astype(int) # Retira parte da string ('(mim)') e converte para inteiro

df['Weatherconditions'] = df['Weatherconditions'].str.replace('conditions ', '')

df.dtypes
```

## Reseta o index do DF

```python
# como excluímos linhas que continham NaN, o index do nosso df ficou com "buracos" (se fizer um df.head(20) na célula a cima, verá que não há index 17)
# os index das linhas excluídas foram excluídos junto

# Devido a isso temos que resetar o index do df para que fique na sequência correta
df = df.reset_index(drop=True) # O drop é necesário pois sem ele o reset cria uma coluna com os index antigos
df.head(20)
```

## Retira todos os espaços desnecessários em colunas do tipo "object"

```python
# Como podemos observar nesse comando, as str dessa tabela apresentam um espaço no fim, oque não é bom!
# Vamos excluir esses espaços das str
df.iloc[0,0]
```

```python
# para excluir espaços usamos o strip()
print(df.iloc[0,0].strip())

# porém precisamos fazer isso em todas as colunas
# usamos FOR para percorrer todas as colunas do tipo Object
for a in [0, 1, 9, 10, 11, 12, 14, 15, 17, 18]:
  df.iloc[:, a] = df.iloc[:, a].str.strip() # Mais rápido que fazer outro FOR

print('Espaços desnecessários excluídos!!!')
  # OU
  #for b in range(0, df.shape[0]):
    #df.iloc[b, a] = df.iloc[b, a].strip()
```

## Cria uma nova coluna, que mostra a distância entre o restaurante e o local de entrega

```python
# Instala a biblioteca usada na célula a baixo, para criar a coluna distância.
!pip show geopy

#!pip install geopy
```

```python
# Adicionando uma nova coluna, que mostra a distância entre o restaurante e o local de entrega
from geopy.distance import great_circle

# Função para calcular a distância usando great_circle
def calcular_distancia(row):
    coord_restaurante = (row['Restaurant_latitude'], row['Restaurant_longitude'])
    coord_entrega = (row['Delivery_location_latitude'], row['Delivery_location_longitude'])
    return great_circle(coord_restaurante, coord_entrega).kilometers

# Aplicar a função ao DataFrame
df['distancia_km'] = df.apply(calcular_distancia, axis=1)

# Exibir o DataFrame com a distância calculada (última coluna)
df
```
